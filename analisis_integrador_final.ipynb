```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
types = {'country':np.int64, 'device_id':np.int64, 'platform': np.int8, 'ref_type_id': np.int8, 'source_id': np.int8}
auctions = pd.read_csv('data/auctions.csv.gzip', dtype = types, parse_dates=['date'])
auctions.head()
```

```python
types = {'event_id':np.int16, 'ref_type':np.int64, 'ref_hash':np.int64, 'aplication_id':np.int16, 'ip_address':np.int64, 'atributed':bool, 'device_countrycode':np.int64,\
        'trans_id':'category', 'device_brand': np.float64, 'device_os':'category', 'wifi':object, 'connection_type':'category'}
events = pd.read_csv('data/events.csv.gzip', dtype=types, parse_dates= ['date'])
```

```python
events.head()
```

```python
types = {'created':'str', 'application_id': 'category', 'ref_type': 'category', 'device_brand': np.float64, \
                'device_countrycode': np.int64, 'device_languaje': 'category'}

installs = pd.read_csv('data/installs.csv.gzip', dtype = types, parse_dates= ['created'])
installs.drop('click_hash', axis=1, inplace=True) # Elimino esta columna por tener todos Nan
```

```python
installs.head()
```

```python
types = {'advertiser_id': np.int8}

clicks = pd.read_csv('data/clicks.csv.gzip', dtype = types, parse_dates= ['created'])
```

```python
clicks.head()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,7))
clicks['specs_brand'].value_counts().nlargest(3).plot(kind='pie', ax=ax1)
installs['ref_type'].value_counts().nlargest(3).plot(kind='pie', ax=ax2)
plt.subplots_adjust(wspace=0.5)
ax1.set_title('Clicks',size=15)
ax2.set_title('Installs',size=15)
plt.show()
```

```python
events_ = events.filter(['date', 'ref_hash', 'attributed'])
clicks_ = clicks.filter(['created', 'timeToClick', 'touchX', 'touchY', 'ref_hash'])
events_ = events_[events_['ref_hash'].isin(clicks_['ref_hash'])]
clicks_with_events = clicks_[clicks_['ref_hash'].isin(events_['ref_hash'])]

def add_next_click_date(ref_hash, date, df):
    date = date.tz_localize(None)
    return df[(df['ref_hash'] == ref_hash) & (df['date'] > date)].min()['date']
    
clicks_with_events['next_click_date'] = clicks_with_events.apply(lambda row: add_next_click_date(row['ref_hash'], row['created'], events_), axis=1)
clicks_with_events = clicks_with_events.dropna(subset=['next_click_date'])
clicks_with_events['time_diff'] = clicks_with_events['next_click_date'] - clicks_with_events['created'].dt.tz_localize(None)
clicks_with_events['time_category'] = pd.cut(clicks_with_events['time_diff'], 
                                             bins=[pd.Timedelta('0'), pd.Timedelta('10 minutes'), pd.Timedelta('30 minutes'), pd.Timedelta('1 hours'), 
                                                   pd.Timedelta('3 hours'), pd.Timedelta('6 hours'), pd.Timedelta('24 hours'), pd.Timedelta('48 hours'), pd.Timedelta('10 days')], 
                                             include_lowest=True, labels=['<10 minutos', '<30 minutos', '<1 hora', '<3 horas', '<6 horas', '<24 horas', '<48 horas', 'Otros'])

with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots(figsize=(11, 16))
plt.title("Dispositivos que realizaron click y luego un evento, \nsegún el intervalo de tiempo entre los sucesos", size=20)

palette = sns.cubehelix_palette(8, dark=.1, light=.9, reverse=True)
sns.scatterplot(x="touchX", y="touchY",
                palette=palette,
                hue="time_category",
                size="time_category",
                alpha=0.9,
                sizes=(30, 180), linewidth=0,
                data=clicks_with_events, ax=ax)
ax.legend().texts[0].set_text("Intervalo tiempo")
ax.set_ylabel("Coordenada en Y", fontsize=18);
ax.set_xlabel("Coordenada en X", fontsize=18);
ax.tick_params(labelsize=15)
plt.setp(ax.get_legend().get_texts(), fontsize='13')
plt.savefig('images/clicks_then_events.png', bbox_inches = 'tight', pad_inches = 0.1)
plt.show()
```

```python
events_ = events.filter(['date', 'ref_hash', 'attributed'])
clicks_ = clicks.filter(['created', 'timeToClick', 'touchX', 'touchY', 'ref_hash'])
events_ = events_[(events_['attributed'] == True) & (events_['ref_hash'].isin(clicks_['ref_hash']))]
clicks_with_events = clicks_[clicks_['ref_hash'].isin(events_['ref_hash'])]

def add_next_click_date(ref_hash, date, df):
    date = date.tz_localize(None)
    return df[(df['ref_hash'] == ref_hash) & (df['date'] > date)].min()['date']
    
clicks_with_events['next_click_date'] = clicks_with_events.apply(lambda row: add_next_click_date(row['ref_hash'], row['created'], events_), axis=1)
clicks_with_events = clicks_with_events.dropna(subset=['next_click_date'])
clicks_with_events['time_diff'] = clicks_with_events['next_click_date'] - clicks_with_events['created'].dt.tz_localize(None)
clicks_with_events['time_category'] = pd.cut(clicks_with_events['time_diff'], 
                                             bins=[pd.Timedelta('0'), pd.Timedelta('5 minutes'), pd.Timedelta('30 minutes'), 
                                                   pd.Timedelta('1 hour'), pd.Timedelta('12 hours'), pd.Timedelta('2 days'), pd.Timedelta('5 days')], 
                                             include_lowest=True, labels=['<5 minutos', '<30 minutos', '<1 hora', '<12 horas', '<2 días', '<5 días'])
with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots(figsize=(11, 16))
plt.title("Dispositivos que realizaron click y luego un evento atribuido a Jampp, \nsegún el intervalo de tiempo entre los sucesos", size=20)

palette = sns.cubehelix_palette(6, dark=.1, light=.8, reverse=True)
sns.scatterplot(x="touchX", y="touchY", palette=palette,hue="time_category",size="time_category", alpha=1 ,sizes=(70, 200), linewidth=0,data=clicks_with_events, ax=ax)
ax.legend().texts[0].set_text("Intervalo tiempo")
ax.set_ylabel("Coordenada en Y", fontsize=18);
ax.set_xlabel("Coordenada en X", fontsize=18);
ax.set(ylim=(-1, 15.9))
ax.tick_params(labelsize=15)
plt.setp(ax.get_legend().get_texts(), fontsize='13')
plt.savefig('images/clicks_then_events_attributed.png', bbox_inches = 'tight', pad_inches = 0.1)
plt.show()
```

```python
auctions_ = auctions.groupby('device_id').filter(lambda x: len(x) > 1).groupby('device_id').agg({'date': ['min', 'max'], 'device_id': 'size'}).reset_index()
auctions_['mean_time'] = (auctions_['date']['max'] - auctions_['date']['min']) / auctions_['device_id']['size']
auctions_.columns = ['device_id', 'min_date', 'max_date', 'appearences', 'mean_time']
auctions_['mean_time_cat'] = pd.cut(auctions_['mean_time'], 
                                             bins=[pd.Timedelta('0'), pd.Timedelta('1 second'), pd.Timedelta('10 seconds'), pd.Timedelta('30 seconds'), 
                                                   pd.Timedelta('1 minute'), pd.Timedelta('5 minutes'), pd.Timedelta('20 minutes'), pd.Timedelta('1 hour'), 
                                                   pd.Timedelta('3 hours'), pd.Timedelta('6 hours'), pd.Timedelta('12 hours'),pd.Timedelta('1 day'),pd.Timedelta('3 days'), pd.Timedelta('10 day')], 
                                                   include_lowest=True, 
                                                   labels=['<1 segundo', '<10 segundos', '<30 segundos', '<1 minuto', '<5 minutos', '<20 minutos', 
                                                           '<1 hora', '<3 horas', '<6 horas', '<12 horas', '<1 dia', '<3 dia','Otros'])
time_data = auctions_.groupby('mean_time_cat').size().reset_index()
time_data.columns = ['mean_time_cat', 'count']

with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots(figsize=(20, 10))
plt.title("Cantidad de dispositivos según cuanto tardan en promedio en volver a aparecer en una subasta", size=22)
palette = sns.cubehelix_palette(13, dark=.1, light=.9, reverse=True)
sns.barplot(x='mean_time_cat', y='count',
        palette=palette,
        data=time_data, ax=ax)
ax.tick_params(labelsize=15)
plt.xticks(rotation=35)
ax.set_ylabel('Cantidad de dispositivos', size=18)
ax.set_xlabel('Tiempo promedio en volver a aparecer en una subasta', size=18)
plt.savefig('images/auctions_mean_appear_time.png', bbox_inches = 'tight', pad_inches = 0.1)
plt.show();
```
