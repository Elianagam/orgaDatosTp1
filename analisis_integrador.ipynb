```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
types = {'country':np.int64, 'device_id':np.int64, 'platform': np.int8, 'ref_type_id': np.int8, 'source_id': np.int8}
auctions = pd.read_csv('data/auctions.csv.gzip', dtype = types, parse_dates=['date'])
#auctions.drop_duplicates(subset = 'device_id', inplace = True)
auctions.head()
```

```python
types = {'event_id':np.int16, 'ref_type':np.int64, 'ref_hash':np.int64, 'aplication_id':np.int16, 'ip_address':np.int64, 'atributed':bool, 'device_countrycode':np.int64,\
        'trans_id':'category', 'device_brand': np.float64, 'device_os':'category', 'wifi':object, 'connection_type':'category'}
events = pd.read_csv('data/events.csv.gzip', dtype=types, parse_dates= ['date'])
```

```python
events.head()
```

```python
types = {'created':'str', 'application_id': 'category', 'ref_type': 'category', 'device_brand': np.float64, \
                'device_countrycode': np.int64, 'device_languaje': 'category'}

installs = pd.read_csv('data/installs.csv.gzip', dtype = types, parse_dates= ['created'])
installs.drop('click_hash', axis=1, inplace=True) # Elimino esta columna por tener todos Nan
```

```python
installs.head()
```

```python
types = {'advertiser_id': np.int8}

clicks = pd.read_csv('data/clicks.csv.gzip', dtype = types, parse_dates= ['created'])
```

```python
clicks.head()
```

```python
mean_clicks = clicks['created'].dt.hour.mean()
```

```python
clicks['created'].max() - clicks['created'].min()
```

```python
mean_auctions = auctions['date'].dt.hour.mean()
```

```python
auctions['date'].max() - auctions['date'].min()
```

```python
mean_events = events['date'].dt.hour.mean()
```

```python
events['date'].max() - events['date'].min()
```

```python
mean_installs = installs['created'].dt.hour.mean()
```

```python
installs['created'].max() - installs['created'].min()
```

```python
labels = ['Clicks', 'Auctions', 'Events', 'Installs']
y_pos = np.arange(len(labels))
mean_hour = [mean_clicks, mean_auctions, mean_events, mean_installs]
 
plt.bar(y_pos, mean_hour, align='center', alpha=0.5)
plt.xticks(y_pos, labels)
plt.ylabel('Hora promedio')
plt.title('Hora promedio de ocurrencia para cada tipo de suceso', size=14)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,7))
clicks['specs_brand'].value_counts().nlargest(3).plot(kind='pie', ax=ax1)
installs['ref_type'].value_counts().nlargest(3).plot(kind='pie', ax=ax2)
plt.subplots_adjust(wspace=0.5)
ax1.set_title('Clicks',size=15)
ax2.set_title('Installs',size=15)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,7))
clicks['carrier_id'].value_counts().nlargest(14).plot(kind='pie', ax=ax1)
events['carrier'].value_counts().nlargest(14).plot(kind='pie', ax=ax2)
plt.subplots_adjust(wspace=0.5)
ax1.set_title('Clicks',size=15)
ax2.set_title('Events',size=15)
plt.show()
```
