```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
types = {'country':np.int64, 'device_id':np.int64, 'platform': np.int8, 'ref_type_id': np.int8, 'source_id': np.int8}
auctions = pd.read_csv('data/auctions.csv.gzip', dtype = types, parse_dates=['date'])
#auctions.drop_duplicates(subset = 'device_id', inplace = True)
auctions.head()
```

```python
types = {'event_id':np.int16, 'ref_type':np.int64, 'ref_hash':np.int64, 'aplication_id':np.int16, 'ip_address':np.int64, 'atributed':bool, 'device_countrycode':np.int64,\
        'trans_id':'category', 'device_brand': np.float64, 'device_os':'category', 'wifi':object, 'connection_type':'category'}
events = pd.read_csv('data/events.csv.gzip', dtype=types, parse_dates= ['date'])
```

```python
top_auctions = auctions['device_id'].value_counts().head(5)
top_clicks = clicks['ref_hash'].value_counts().head(5)

newdf = auctions
newdf2 = clicks
newdf['hour'] = newdf['date'].dt.hour
newdf['day'] = newdf['date'].dt.date
newdf2['hour'] = newdf2['created'].dt.hour
newdf2['day'] = newdf2['created'].dt.date

days = list(clicks.drop_duplicates('day')['day'].sort_values())

for i in range(len(top_auctions.index)):
    fig, axs = plt.subplots(1, 2, figsize=(20,7))
    titles = ['Dia', 'Hora']
    labels = ['Auctions', 'Clicks']
    column_names = ['day', 'hour']
    fig.suptitle('Top '+str(i+1)+' en auctions y en clicks', size=16)
    for j in range(2):
        ax = axs[j]
        ax_twin = ax.twinx()
        width = 0.4
        newdf2 = clicks[clicks['ref_hash'] == top_clicks.index[i]]
        newdf2.groupby(column_names[j]).size().plot(kind='bar', color='blue', ax=ax_twin, position=0, width=width, label='Clicks')
        ax_twin.legend(loc=1)
        ax_twin.set_ylabel(labels[1])
        ax_twin.set_xlabel(titles[1])
        newdf = auctions[auctions['device_id'] == top_auctions.index[i]]
        newdf.groupby(column_names[j]).size().plot(kind='bar', color='red', ax=ax, position=1, width=width, label='Auctions')
        ax.legend(loc=2)
        ax.set_ylabel(labels[0])
        ax.set_xlabel(titles[0])
    plt.subplots_adjust(wspace=0.5)
    plt.show()
```

```python
fig, ax = plt.subplots()
newdf = clicks.groupby('ref_hash').size().sort_values(ascending=False).reset_index()
newdf.columns = ['ref_hash', 'count']
newdf_ = newdf.groupby('count').size().reset_index()
newdf_.columns = ['appearences', 'devices']

newdf_.plot(kind='scatter', ax=ax, x='appearences', y='devices', loglog=False, alpha=0.5, figsize=(14,10))
plt.title('Cantidad de dispositivos para cada cantidad de apariciones de subastas', size=16)
ax.set_xlabel('Cantidad de apariciones')
ax.set_ylabel('Cantidad de dispositivos')
plt.show()
```

```python
events_ = events.filter(['date', 'ref_hash', 'attributed'])
clicks_ = clicks.filter(['created', 'timeToClick', 'touchX', 'touchY', 'ref_hash'])
```

```python
events_ = events_[events_['ref_hash'].isin(clicks_['ref_hash'])]
clicks_with_events = clicks_[clicks_['ref_hash'].isin(events_['ref_hash'])]
```

```python
with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots(figsize=(10, 18))
    ax.set_title("Clicks en la pantalla", fontsize=20)
    
sns.scatterplot(x="touchX", y="touchY",
                color='g',
                alpha=0.4,
                linewidth=0,
                s=40,
                data=clicks_with_events, ax=ax)
ax.set_ylabel("Coordenada en Y", fontsize=15)
ax.set_xlabel("Coordenada en X", fontsize=15)
ax.tick_params(labelsize=10, pad=10)
```

```python
def add_next_click_date(ref_hash, date, df):
    date = date.tz_localize(None)
    return df[(df['ref_hash'] == ref_hash) & (df['date'] > date)].min()['date']
    
clicks_with_events['next_click_date'] = clicks_with_events.apply(lambda row: add_next_click_date(row['ref_hash'], row['created'], events_), axis=1)
clicks_with_events = clicks_with_events.dropna(subset=['next_click_date'])
clicks_with_events['time_diff'] = clicks_with_events['next_click_date'] - clicks_with_events['created'].dt.tz_localize(None)
clicks_with_events['time_category'] = pd.cut(clicks_with_events['time_diff'], 
                                             bins=[pd.Timedelta('0'), pd.Timedelta('10 minutes'), pd.Timedelta('30 minutes'), pd.Timedelta('1 hours'), 
                                                   pd.Timedelta('3 hours'), pd.Timedelta('6 hours'), pd.Timedelta('24 hours'), pd.Timedelta('48 hours'), pd.Timedelta('7 days')], 
                                             include_lowest=True, labels=['10 minutos', '30 minutos', '1 hora', '3 horas', '6 horas', '24 horas', '48 horas', 'Otros'])
```

```python
# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots(figsize=(10, 18))
plt.title("Clicks que resultaron en eventos segun el intervalo de tiempo entre el click y evento", size=18)
palette = sns.cubehelix_palette(8, dark=.1, light=.9, reverse=True)
sns.scatterplot(x="touchX", y="touchY",
                palette=palette,
                hue="time_category",
                size="time_category",
                alpha=0.8,
                sizes=(20, 180), linewidth=0,
                data=clicks_with_events, ax=ax)
```
